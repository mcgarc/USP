# Zeeman guiding example analysis

Run Jupyter with USP requirements installed!

```python
import sys
sys.path.append('..')
from USP import *

import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

Run `python zeeman.py` to populate results directory with the pickled simulation instance. This is ok here because we only do a few particles, but if you have a lot then it is better to use the sampling functionality.

This pickle can then be loaded

```python
sim = simulation.load_pickle('../results/zeeman_example.pickle')
```

Create times array and get the mean position of the particles at each point in time.

```python
ts = np.linspace(0, sim.t_end, 100)
rs = [[p.r(t) for p in sim._particles] for t in ts]
mean_rs = np.mean(rs, axis=1)
```

Plot mean x

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(ts, 1000*mean_rs[:,0])
```

Plot mean y

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(ts, 1000*mean_rs[:,1])
```

Plot mean z

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(ts, 1000*mean_rs[:,2])
```

Look at the temperature across time. Note that temperature is not energy of the system here, since some energy is stored in the position of the particles

```python
temps = [sim.temperature(t) for t in ts]
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(ts, temps)
```
